# AutoML y Procesamiento del Lenguaje Natural (NLP)

## Pycaret

PyCaret es una herramienta de automatización del flujo de trabajo para el aprendizaje automático supervisado y no supervisado.

Todos los módulos en PyCaret admiten la preparación de datos (más de 25 técnicas de preprocesamiento esenciales, ajuste automático de hiperparámetros, análisis e interpretación de modelos, selección automática de modelos, registro de experimentos y opciones de implementación en la nube)


## Procesamiento de lenguaje natural

El procesamiento del lenguaje natural (PNL) es un subcampo de la informática y la inteligencia artificial que se ocupa de las interacciones entre las computadoras y los lenguajes humanos. En particular, la PNL cubre una amplia gama de técnicas sobre cómo programar computadoras para procesar y analizar grandes cantidades de datos en lenguaje natural.

### ¿Que es el Topic Modeling?

Un modelo de tema es un tipo de modelo estadístico que se incluye en el aprendizaje automático no supervisado y se utiliza para descubrir temas abstractos en datos de texto. El objetivo del modelado de temas es encontrar automáticamente los temas / temas en un conjunto de documentos.
Algunos casos de uso comunes para el modelado de temas son:
- Resumir datos de texto grandes clasificando documentos en temas ( la idea es bastante similar a la agrupación en clústeres ).
- Análisis de datos exploratorios para comprender datos como formularios de comentarios de clientes, reseñas de Amazon, resultados de encuestas, etc.
- Ingeniería de funciones que crea funciones para experimentos de aprendizaje automático supervisados, como clasificación o regresión

Se utilizan varios algoritmos para el modelado de temas. Algunos de los más comunes son la asignación de Dirichlet latente (LDA), el análisis semántico latente (LSA) y la factorización de matriz no negativa (NMF).

### Procesamiento previo de texto para modelado de temas

Para obtener resultados significativos del modelado de temas, los datos de texto deben procesarse antes de introducirlos en el algoritmo. Esto es común con casi todas las tareas de PNL. El preprocesamiento de texto es diferente de las técnicas clásicas de preprocesamiento que se utilizan a menudo en el aprendizaje automático cuando se trata de datos estructurados (datos en filas y columnas).

PyCaret automáticamente preprocesa los datos de texto mediante la aplicación de más de 15 técnicas, tales como la **eliminación de palabras vacías , tokenización , lematización , bi-gramo / tri-gramo extracción** etc

# Comandos básicos de NLP en Pycaret

Pycaret contiene numerosas funciones y atributos. Las funciones que se utilizan en la mayoria de casos son las siguientes:

1. <code>create_model </code>crea el modelo de NLP

2. <code> setup() </code>establece el entorno de trabajo.

3. <code>assign_model </code>se utiliza para asignar-predecir el topic.

4. <code>plot_model </code> se utiliza para evaluar el rendimiento del modelo de aprendizaje automático entrenado.
